### Encoding the categorical data

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
data['Category'] = LE.fit_transform(data['Category'])
data['Sex'] = LE.fit_transform(data['Sex'])

data
```

### To see the distribution of the numerical values and check if they need to be normalized

```python
data.describe()
```

### Scaling numerical values

```python
from sklearn.preprocessing import MinMaxScaler

columns_names = data.columns.tolist()
columns_to_scale = columns_names
columns_to_remove = ['Unnamed: 0', 'Category', 'Sex']
for column in columns_to_remove: columns_to_scale.remove(column)

scaler = MinMaxScaler()

for column in columns_to_scale:
    data[column] = scaler.fit_transform(data[[column]])
```

```python
data
```

> Data types are all numeric Now ! Next we need to deal with missing values for the feature columns 

#To-Do list all the columns that contain missing values along with their counts 

```python
# To-Do Start providing solutions for each column that has missing data 
# Treat each case seperately 
# Hint : no data row should be deleted 
# Provide evidence that you filled those missing values after each step 
```

```python
data.isnull().sum()
```

```python
# Imputing values for columns with single missing value

data['ALB'] = data['ALB'].fillna(data['ALB'].mean())
data['ALT'] = data['ALT'].fillna(data['ALT'].median())
data['PROT'] = data['PROT'].ffill()

# Interpolating missing values

data['CHOL'] = data['CHOL'].interpolate()

# Using predictive imputation K-means

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
data['ALP'] = imputer.fit_transform(data[['ALP']])

data.isnull().sum()
```

> We have no more missing data
